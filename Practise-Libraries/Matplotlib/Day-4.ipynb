# Day 4 - Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Visualization Styles

````python
np.random.seed(28381)
data1, data2, data3, data4 = np.random.randn(4, 100)

fig, ax = plt.subplots(figsize=(5, 2.8))
x = np.arange(len(data1))

# Implicity set the properties of the graph
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle='--')

# Explicitly set the properties of the graph
l, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=5)
l.set_linestyle(':')
````

### Color
```
'b' as blue
'g' as green
'r' as red
'c' as cyan
'm' as magenta
'y' as yellow
'k' as black
'w' as white
```

````python
fig, ax = plt.subplots(figsize=(5, 2.8))
ax.scatter(data1, data2, s=50, facecolor='r', edgecolors='b')
````

### Linestyle

Named linestyles
```
'solid'
'dotted'
'dashed'
'dashdot'
```
Parametrized linestyles
```
(0, (1, 10)) -> 'loosely dotted'
(0, (1, 1))  -> 'dotted'
(0, (1, 1))  -> 'densely dotted
(5, (10, 3)) -> 'long dash with offset'
```

### Marker Style
```
'o' -> Circle style
'd' -> Diamonds style
'v' -> Inverted triangle
's' -> Square
```

````python
fig, ax = plt.subplots(figsize=(5, 2.8))

ax.plot(data1, 'o', label='data1')
ax.plot(data2, 'd', label='data2')
ax.plot(data3, 'v', label='data3')
ax.plot(data4, 's', label='data4')

ax.legend()
````
