# Day 5 - Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Labeling Plots

### Common labels and texts and Mathematical symbols

#### Histogram

```python
np.random.seed(1968)

mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)

fig, ax = plt.subplots(figsize=(5.8, 3), constrained_layout=True)

# Histrogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.8)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Anteator lengths')

ax.text(150, 0.015, 'Random Text')

# Mathematical symbols
ax.text(100, 0.02, r'$\sigma_i=15$')
ax.text(80, 0.01, r'$\alpha > \beta$')
ax.text(75, 0.025, r'$\mu=115\ \sigma=15$')

ax.grid(True)
```

### Annotations

```python
fig, ax = plt.subplots(figsize=(5.8, 3))
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```

### Legends

```python
data1, data2, data3 = np.random.randn(3, 100)

fig, ax = plt.subplots(figsize=(5.8, 3))

x = np.arange(len(data1))

ax.plot(x, data1, 's', label='data1')
ax.plot(x, data2, label='data2')
ax.plot(x, data3, 'd', label='data3')

ax.legend()
```
