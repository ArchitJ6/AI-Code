# Day 6 - Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scales

```python
np.random.seed(19680801)
fig, axs = plt.subplots(1,2,figsize=(5.8, 3), constrained_layout=True)
data1 = np.random.randn(100)

x = np.arange(len(data1))
y = 10**data1

# length is the 1st column and width is the 2nd column
axs[0].plot(x, y)
axs[1].set_yscale('log')
axs[1].plot(x, y)
```

## Ticks

```python
np.random.seed(15)
fig, axs = plt.subplots(2,1,figsize=(5.8, 3), constrained_layout=True)
data1 = np.random.randn(100)

x = np.arange(len(data1))
axs[0].plot(x, data1)
axs[0].set_title('Automatic ticks')

axs[1].plot(x, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', '60', 'ninety'])
axs[1].set_title('Manual ticks')
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout=True)

axs[0].plot(x, data1)
axs[0].set_title('Automatic ticks')

axs[1].plot(x, data1)
axs[1].set_yticks([-1.5, 0, 1.5]) # Note that we don't need to specity labels
axs[1].set_title('Manual ticks')
```
