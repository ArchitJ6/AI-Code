```python
import pandas as pd
import numpy as np
```

### 1. The df DataFrame is given below. Extract time (in hours) from the timestamp column and assign to the new column 'hour'. Print the first five rows of this DataFrame to the console.

```python
url = 'https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/london_bike.csv'
df = pd.read_csv(url)
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Start your code here
df['hour'] = df.timestamp.dt.hour
df.head()
```

---

### 2. The df DataFrame is given below. Extract month from the timestamp column and assign to the new column 'month'. Print the first five rows of this DataFrame to the console.

```python
df['month'] = df.timestamp.dt.month
df.head()
```

---

### 3. The df DataFrame is given below. Group this DataFrame by month and calculate the average value for the hum column. Assign the result to humidity_by_month variable and reset the index. In response, print humidity_by_month DataFrame to the console as shown below.

```python
humidity_by_month = df.groupby('month')['hum'].mean()
humidity_by_month = humidity_by_month.reset_index()
humidity_by_month
```

---

### 4. The df DataFrame is given below. Group this DataFrame by hour and calculate the average value of the cnt column. Assign the result to the cnt_by_hour variable and reset the index. In repsonse, print cnt_by_hour DataFrame to the console as shown below.

```python
cnt_by_hour = df.groupby('hour')['cnt'].mean().reset_index()
cnt_by_hour
```

---

### 5. The df DataFrame is given below. Group this DataFrame by is_weekend and hour columns and calculate the average value for the cnt variable. Assign the result to the cnt_by_weekend_hour variable and reset the index. In response, print cnt_by_weekend_hour DataFrame to the console as shown below.

```python
cnt_by_weekend_hour = df.groupby(['is_weekend', 'hour'])['cnt'].mean().reset_index()
cnt_by_weekend_hour
```
