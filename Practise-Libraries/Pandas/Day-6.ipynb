### 1. The google.csv file is loaded using pandas as shown below. Find the index for which the Close price has the lowest value and assign it to the idx_min variable. Then extract the row for this index from the google DataFrame as a DataFrame. In response, print this DataFrame to the console.

```python
google = pd.read_csv('google.csv', index_col=0)
google = google.reset_index()

# Start your code here
idx_min = google.Close.argmin()
print(google.iloc[[idx_min]])
```

### 4. The google.csv file is loaded using pandas as shown below. Remove the Year and Month columns from the google DataFrame. In response, print google DataFrame to the console.

```python
google = pd.read_csv('google.csv', index_col=0)
google['Date'] = pd.to_datetime(google['Date'])
google['Year'] = google['Date'].dt.year
google['Month'] = google['Date'].dt.month

# Start your code here
google = google.drop(columns=['Year', 'Month'])
google
```
